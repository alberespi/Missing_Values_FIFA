## **MCAR GENERATION**

```python
def generate_mcar(data, pr):
    x = len(data)
    y = len(data.columns)
    num_missing_values = int(x * y * pr)

    missing_row_indices = np.random.randint(0, x, num_missing_values)
    missing_col_indices = np.random.randint(0, y, num_missing_values)

    for i in range(num_missing_values):
        data.iloc[missing_row_indices[i], missing_col_indices[i]] = np.nan
    
    return data
```

```python
probs = [0.1, 0.3, 0.5]

mcar_dfs = {}

for p in probs:
    mcar = df.copy()
    mcar_dfs[f'mcar_df_{int(p*100)}'] = generate_mcar(mcar, p)
```

## **MAR GENERATION**

```python
def generate_mar(data, pr):
    x = len(data)
    y = len(data.columns)
    num_missing_values = int(x * y * pr)

    # Calculate the correlation matrix and fill NaN values with 0
    corr_matrix = data.corr().abs().fillna(0)

    # Get the probability weights for each column based on their maximum correlation
    prob_weights_columns = corr_matrix.max()

    for _ in range(num_missing_values):
        # Choose a column based on its probability weight
        selected_column = np.random.choice(
            data.columns, p=prob_weights_columns / prob_weights_columns.sum()
        )

        # Calculate the probabilities for selecting a row, handling possible NaN values
        row_probabilities = data[selected_column].fillna(0) / data[selected_column].sum()

        # If all probabilities are zero, set them to uniform probabilities
        if row_probabilities.sum() == 0:
            row_probabilities = pd.Series([1 / len(row_probabilities)] * len(row_probabilities), index=row_probabilities.index)

        # Normalize the probabilities so that they sum to 1
        row_probabilities /= row_probabilities.sum()

        # Choose a row based on the correlation between the selected column and other columns
        selected_row = np.random.choice(
            data.index,
            p=row_probabilities,
        )

        # Set the value to NaN
        data.at[selected_row, selected_column] = np.nan

    return data

```

In the MAR missing value generation algorithm, the correlation matrix is used to simulate the "Missing At Random" pattern. Missing At Random (MAR) means that the probability of a value being missing is related to the observed data. In other words, the missingness is not entirely random, but it is also not dependent on the specific value that is missing.

The correlation matrix is a measure of the linear relationship between pairs of variables in the dataset. By using the correlation matrix, the algorithm can identify relationships between variables and exploit them to introduce missing values. This way, the algorithm simulates the MAR pattern more accurately.

In the algorithm, we calculate the correlation matrix and obtain the maximum correlation for each column. These maximum correlations are used as probability weights for selecting which variable (column) to introduce missing values. A higher correlation indicates a stronger relationship between variables, making it more likely for the variable to be selected for missing value introduction. Once a column is selected, a row is chosen based on the distribution of the selected column, which also reflects the relationships between variables.

By using the correlation matrix, the algorithm ensures that the missing values are introduced in a way that mimics the MAR pattern – the probability of missing values depends on the observed data, capturing the relationships between variables in the dataset.

```python
mar_dfs = {}

for p in probs:
    mar = df.copy()
    mar_dfs[f'mar_df_{int(p*100)}'] = generate_mar(mar, p)
```

## **MNAR GENERATION**

```python
def generate_mnar(data, pr):
    x = len(data)
    y = len(data.columns)
    num_missing_values = int(x * y * pr)

    data_flat = data.values.flatten()
    data_flat_indices = np.arange(data_flat.size)
    prob_weights = np.abs(np.random.randn(data_flat.size))

    missing_indices = np.random.choice(
        data_flat_indices, num_missing_values, replace=False, p=prob_weights / prob_weights.sum()
    )

    for idx in missing_indices:
        i, j = divmod(idx, y)
        data.iloc[i, j] = np.nan

    return data
```

```python
mnar_dfs = {}

for p in probs:
    mnar = df.copy()
    mnar_dfs[f'mnar_df_{int(p*100)}'] = generate_mnar(mnar, p)
```

```python
print(mnar_dfs['mnar_df_10'].isna().sum().sum())
print(mnar_dfs['mnar_df_30'].isna().sum().sum())
print(mnar_dfs['mnar_df_50'].isna().sum().sum())
```

In the MNAR (Missing Not At Random) missing value generation algorithm, the missingness is related to the values that are missing themselves, rather than to the observed data. To simulate this type of missingness, the algorithm introduces a probability weight for each data point, which is based on a random noise distribution (in this case, a standard normal distribution).

Here's a breakdown of the MNAR algorithm:

Flatten the input DataFrame into a 1D array (data_flat) and calculate the indices of each element (data_flat_indices).
Generate a noise distribution by drawing random samples from a standard normal distribution (using np.random.randn()). This noise distribution is the same size as the flattened data.
Take the absolute value of the noise distribution to ensure positive weights and use them as probability weights (prob_weights) for each data point.
Choose a certain number of indices (missing_indices) from the flattened data based on the given percentage (p) of missing values and the probability weights.
Replace the selected elements in the original DataFrame with np.nan to introduce missing values.
In this algorithm, the probability of a value being missing depends on the random noise distribution. Since the missingness is generated using a random noise distribution unrelated to the observed data, the missingness is more likely to be related to the missing values themselves rather than the observed values. This approach simulates the MNAR pattern in the dataset.

It's important to note that this specific MNAR algorithm is just one way to introduce MNAR missing values. Depending on the desired missing data pattern or specific use case, other approaches might be more suitable.
