```python
# Estadisticas basicas de las columnas numericas
df.describe()
```

```python
# Grafico de histograma para visualizar la distribucion de la media de todos los jugadores
df["Overall"].plot.hist(bins=50)
plt.xlabel("Overall")
plt.ylabel("Frequency")
plt.show()
```

```python
# Grafico de caja para visualizar la distribucion de la edad de todos los jugadores
sns.violinplot(x=df['Age'])
plt.xlabel("Age")
plt.show()
```

```python
# Grafico de dispersion para visualizar la relacion entre la media actual del jugador con su respectivo
# potencial
sns.scatterplot(x=df["Overall"], y=df["Potential"])
plt.xlabel("Overall")
plt.ylabel("Potential")
plt.show()
```

```python
corr = df[["Overall", "Potential", "Age", "TotalStats", "BaseStats", "Pace Total", "Shooting Total",
           "Passing Total", "Dribbling Total", "Defending Total", "Physicality Total"]].corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
df.columns
```

```python
df['Best Position'].value_counts().plot(kind='bar')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.show()
```

```python
df_copy = df.copy()
nationalities = df_copy["Nationality"].value_counts(normalize=True)

other_nationalities = nationalities[nationalities < 0.015].index

df_copy['Nationality'] = df_copy['Nationality'].replace(other_nationalities, "Other")

nationalities = df_copy['Nationality'].value_counts(normalize=True)
nationalities.plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
sns.kdeplot(df['Overall'])
plt.xlabel('Overall')
plt.ylabel('Density')
plt.show()
```

```python
top9_nationalities = ['Brazil', 'Spain', 'England', 'France', 'Italy', 'Argentina', 'Germany', 'Netherlands', 'Portugal', 'Belgium']

df_copy = df[df["Nationality"].isin(top9_nationalities)]

sns.heatmap(pd.crosstab(df_copy['Best Position'], df_copy['Nationality']),
            cmap="YlGnBu", annot=True, fmt='d')
plt.xlabel('Nationality')
plt.ylabel('Position')
plt.show()
```
