```python
df['Best Position'].value_counts().plot(kind='bar')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.show()
```

```python
df_copy = df.copy()
nationalities = df_copy["Nationality"].value_counts(normalize=True)

other_nationalities = nationalities[nationalities < 0.025].index

df_copy['Nationality'] = df_copy['Nationality'].replace(other_nationalities, "Other")

nationalities = df_copy['Nationality'].value_counts(normalize=True)
nationalities.plot.pie(autopct='%1.1f%%', pctdistance=0.85, startangle=90, counterclock=False, textprops=dict(color="black"))
plt.show()
```

```python

df_copy = df.copy()
nationalities = df_copy["Nationality"].value_counts(normalize=True)

other_nationalities = nationalities[nationalities < 0.015].index

df_copy['Nationality'] = df_copy['Nationality'].replace(other_nationalities, "Other")

nationalities = df_copy['Nationality'].value_counts(normalize=True)

# Set explode values to separate slices
explode_values = np.zeros(len(nationalities))
explode_values[:] = 0.05

ax = nationalities.plot.pie(autopct='%1.1f%%', pctdistance=0.85, startangle=90, counterclock=False, textprops=dict(color="black"), explode=explode_values)

# Increase spacing between labels
for i, (label, value) in enumerate(zip(nationalities.index, nationalities.values)):
    angle_rad = np.deg2rad((360 * value / 2) + (360 * np.sum(nationalities.values[:i])))
    angle_deg = np.rad2deg(angle_rad)
    x_pos = 1.3 * np.cos(angle_rad)
    y_pos = 1.3 * np.sin(angle_rad)

    if 90 < angle_deg <= 270:
        alignment = 'right'
        x_pos = x_pos - 0.05
    else:
        alignment = 'left'
        x_pos = x_pos + 0.05

    ax.annotate(label, xy=(x_pos, y_pos), ha=alignment, va='center', fontsize=10)

plt.ylabel('')  # Remove the ylabel
plt.show()


```
