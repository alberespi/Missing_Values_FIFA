```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/fifa23.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.fillna(df.mean(), inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(columns={'Positions Played', 'Image Link', 'Release Clause', 'Club Position',
                 'Contract Until', 'Club Jersey Number', 'Joined On', 'On Loan', 
                 'National Team Name', 'National Team Image Link', 'National Team Position', 
                 'National Team Jersey Number'}, inplace=True)
```

```python
df.columns
```

```python
non_numeric_columns = df.select_dtypes(exclude='number').columns
print(non_numeric_columns)
```

```python
unique_values_pf = df['Preferred Foot'].unique()
print(unique_values_pf)
```

```python
df["Preferred Foot"].replace({"Left": 0, "Right": 1}, inplace=True)
```

```python
unique_values_pf = df['Preferred Foot'].unique()
print(unique_values_pf)
```

```python
df_encoded = pd.get_dummies(df, columns=["Attacking Work Rate", "Defensive Work Rate"])
df_encoded
```

```python
df_encoded.columns
```

```python
df_categorical = df_encoded.select_dtypes(exclude=['number'])
df_numeric = df_encoded.select_dtypes(include='number')
```

```python
df_categorical
```

```python
df_numeric
```

```python
df_categorical.to_csv('./data/fifa23_cat_clean.csv', index=False)
df_numeric.to_csv('./data/fifa23_num_clean.csv', index=False)
```
