```python
# Generacion de missing values MAR
strat_var = "Age"

mar_mask = df[strat_var] == 25
df_mar = df.mask(mar_mask)

df_mar
```

```python
def miv_generating_algorithm(data, mdp, nV):
    x = len(data)
    y = len(data.columns)
    observations = []
    causatives = np.random.choice(range(y), nV, replace=False)

    for causative in causatives:
        aux = data.iloc[:, causative].copy()
        num_missing_values = int(x * y * mdp / 100 // nV)

        for _ in range(num_missing_values):
            min_index = aux.idxmin()
            observations.append(min_index)
            aux.loc[min_index] = np.iinfo(np.int32).max

    for observation in observations:
        data.iloc[observation, causative] = np.nan

    return data
```

```python
mdp = 10
nV = 2
test = df.copy()
test_mar = miv_generating_algorithm(test, mdp, nV)

test_mar
```

```python
# Generacion de missing values MNAR
cols = df.columns.to_list()
print(cols)
print(len(cols))
```

```python
binomial = ['Attacking Work Rate_High', 'Attacking Work Rate_Low',
            'Attacking Work Rate_Medium', 'Defensive Work Rate_High',
            'Defensive Work Rate_Low', 'Defensive Work Rate_Medium']
normal = list(set(cols) - set(binomial))
print(len(normal))
```

```python
df_mnar = df.copy()


binomial_missing_prob = 0.1

normal_missing_prob = lambda x: 1 - norm.cdf(x, loc=0.5, scale=0.1)

for col in binomial:
    df_mnar[col] = df_mnar[col].apply(lambda x: np.nan if np.random.binomial(1, binomial_missing_prob) else x)

for col in normal:
    min_value = df_mnar[col].min()
    max_value = df_mnar[col].max()

    df_mnar[col] = df_mnar[col].apply(lambda x: (x - min_value) / (max_value - min_value))
    
    df_mnar[col] = df_mnar[col].apply(lambda x: np.nan if np.random.binomial(1, normal_missing_prob(x)) else x)

df_mnar
```
