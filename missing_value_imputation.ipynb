```python
import pandas as pd
import numpy as np
import random
from sklearn.impute import SimpleImputer
random.seed(0)
```

```python
df_mar = pd.read_csv('./data/fifa23_MAR.csv')
df_mcar = pd.read_csv('./data/fifa23_MCAR.csv')
df_mnar = pd.read_csv('./data/fifa23_MNAR.csv')
```

# **MEAN IMPUTATION**

```python
mean = SimpleImputer(missing_values=np.nan, strategy='mean')

mean.fit(df_mar)
df_mar_mean_imputed = mean.transform(df_mar)
df_mar_mean_imputed = pd.DataFrame(df_mar_mean_imputed, columns=df_mar.columns)

mean.fit(df_mcar)
df_mcar_mean_imputed = mean.transform(df_mcar)
df_mcar_mean_imputed = pd.DataFrame(df_mcar_mean_imputed, columns=df_mar.columns)

mean.fit(df_mnar)
df_mnar_mean_imputed = mean.transform(df_mnar)
df_mnar_mean_imputed = pd.DataFrame(df_mnar_mean_imputed, columns=df_mar.columns)
```

```python
df_mnar_mean_imputed
```

# **MOST FREQUENT IMPUTATION**

```python
most_frequent = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

most_frequent.fit(df_mar)
df_mar_mf_imputed = most_frequent.transform(df_mar)
df_mar_mf_imputed = pd.DataFrame(df_mar_mf_imputed, columns=df_mar.columns)

most_frequent.fit(df_mcar)
df_mcar_mf_imputed = most_frequent.transform(df_mcar)
df_mcar_mf_imputed = pd.DataFrame(df_mcar_mf_imputed, columns=df_mar.columns)

most_frequent.fit(df_mnar)
df_mnar_mf_imputed = most_frequent.transform(df_mnar)
df_mnar_mf_imputed = pd.DataFrame(df_mnar_mf_imputed, columns=df_mar.columns)
```
